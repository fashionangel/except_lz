```python
def compare_dataframe_structure(df, expected_columns, expected_dtypes):
    """
    Сравнивает структуру датафрейма: порядок/названия столбцов и их типы.
    
    Параметры:
      df – загруженный датафрейм;
      expected_columns – список ожидаемых названий столбцов в нужном порядке;
      expected_dtypes – словарь с ожидаемыми типами данных для каждого столбца (например, {"id": "int64", "name": "object"}).
      
    Возвращает:
      Список ошибок, обнаруженных при сравнении. Если список пуст – структуры совпадают.
    """
    errors = []
    
    # Сравнение названий столбцов (с учетом порядка)
    actual_columns = list(df.columns)
    if actual_columns != expected_columns:
        errors.append(
            f"Названия столбцов не совпадают.\nОжидаемые: {expected_columns}\nФактические: {actual_columns}"
        )
    
    # Сравнение типов данных для каждого столбца
    for col, expected_type in expected_dtypes.items():
        if col not in df.columns:
            errors.append(f"Отсутствует обязательный столбец: '{col}'")
        else:
            actual_type = str(df[col].dtype)
            if actual_type != expected_type:
                errors.append(
                    f"В столбце '{col}' тип данных не соответствует ожидаемому.\nОжидается: {expected_type}, Фактически: {actual_type}"
                )
    
    return errors




a = "Адрес среда кума мука"

def anagram(text: str) -> list:
    anagram_groups = {}
    words = text.split()

    for word in words:
        # Создаем ключ из отсортированных букв слова
        sorted_letters = tuple(sorted(word.lower()))
        
        # Если ключа еще нет в словаре, создаем для него пустой список
        if sorted_letters not in anagram_groups:
            anagram_groups[sorted_letters] = []
            
        # Добавляем слово в список этой группы
        anagram_groups[sorted_letters].append(word)
        
    # Возвращаем списки слов из значений словаря
    return list(anagram_groups.values())

print(anagram(a))
```
